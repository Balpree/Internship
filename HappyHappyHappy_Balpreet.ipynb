```python
#let's check How Happy is Spain Among its Neighbors?
top = happiness[happiness['Country'] == 'Finland']
south_asia = happiness[happiness['Region'] == 'Western Europe']
top_bottom_neighbors = pd.concat([top, south_asia], axis=0)
top_bottom_neighbors['Rank'] = top_bottom_neighbors.index + 1
top_bottom_neighbors.reset_index(drop=True, inplace=True)
top_bottom_neighbors.drop(['Region', 'Dystopia Residual'], axis=1, inplace=True)
df_glob = happiness.sort_values('Happiness Score', ascending=False)[['Country', 'Happiness Score', 'Region']].reset_index(drop=True)
top = df_glob[df_glob['Happiness Score'] == df_glob['Happiness Score'].max()]
neighbors = df_glob[df_glob['Region'] == 'Western Europe']
top_bottom_neighbors = pd.concat([top, neighbors], axis=0)
top_bottom_neighbors['Rank'] = list(top_bottom_neighbors.index + 1)
top_bottom_neighbors.reset_index(drop=True, inplace=True)
top_bottom_neighbors.drop('Region', axis=1, inplace=True)
mean_score = happiness['Happiness Score'].mean()
rank = list(top_bottom_neighbors['Rank'])
fig, ax = plt.subplots(figsize=(12, 6))
bar = sns.barplot(x='Country', y='Happiness Score', data=top_bottom_neighbors, palette='gist_earth_r')
sns.set_style('whitegrid')
sns.despine(left=True)
for i in range(1, len(top_bottom_neighbors)):
    bar.text(x=i, y=(top_bottom_neighbors['Happiness Score'][i])/2, s=str(rank[i])+'th',
            fontdict=dict(color='white', fontsize=12, fontweight='bold', horizontalalignment='center'))
bar.text(x=0, y=(top_bottom_neighbors['Happiness Score'][0])/2, s=str(rank[0])+'st', fontdict=dict(color='white', fontsize=12, fontweight='bold', horizontalalignment='center'))
bar.axhline(mean_score, color='grey', linestyle='--')
bar.text(x=len(top_bottom_neighbors)-0.4, y = mean_score, s = 'GlobalnAverage: {:.2f}'.format(mean_score),
        fontdict = dict(color='white', backgroundcolor='grey', fontsize=10, fontweight='bold'))
plt.title('How Happy is Spain Among its Neighbors?', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('')
plt.show()
```

Let's prepare the data by splitting the dataset in axis X and axis Y so that we can analyze the impact of the independent feature over the Happiness Score.
Furthermore, let's use dummies in order to convert all the categorical features in 0 AND 1 (dummies variables) and we use  Labelencoder for encoding target labels with value between 0 and n_classes-1
And finally, we use Standard Scaler in order to normalize the data and Standardize features by removing the mean and scaling to unit variance.

```python
X = happiness.drop(['Happiness Rank', 'Happiness Score'], axis = 1)
Y = happiness['Happiness Score']
X = pd.get_dummies(X, prefix_sep='_')
Y = LabelEncoder().fit_transform(Y)
X = StandardScaler().fit_transform(X)
```

We will use Principle Components Analysis (PCA)  for Feature Extraction:
PCA is one of the most used linear dimensionality reduction technique. For PCA, we take our original data and try to find a combination of the input features which can best summarize the original data distribution so that to reduce its original dimensions. PCA is able to do this by maximizing variances and minimizing the reconstruction error by looking at pair wised distances.

In this example, I will first perform PCA in the whole dataset to reduce our data to just four dimensions and I will then construct a data frame with our new features and their respective labels.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)
X_pca = pca.fit_transform(X)
PCA_df = pd.DataFrame(data = X_pca, columns = ['PC1', 'PC2', 'PC3','PC4'])
PCA_df = pd.concat([PCA_df, happiness['Happiness Score']], axis = 1)
PCA_df['Happiness Score'] = LabelEncoder().fit_transform(PCA_df['Happiness Score'])
PCA_df.head()
```
