```python
#getting the data from the website
Ranks = []
Names = []
Artists = []
Upload_date=[]
views=[]


        #Ranks
    
    #getting the tags having the ranks
try:
    brand_tgs=driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[1]")
except:
    Ranks.append('-')

#recolecting the ranks
Ranks=[]
for i in brand_tgs:
    Ranks.append(i.text)
Ranks
    
    #getting the tags having the names


        #Names
try:
    desc_name=driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[2]")
    Names.append(desc_name.text)
except:
    Names.append('-')
    
    #getting the tags having the names
Names=[]
for i in desc_name:
    Names.append(i.text)
Names

        #Artists
try:
    rating_tg=driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[3]")
    Artists.append(rating_tg.text)
except:
    Artists.append('-')

Artists=[]
for i in rating_tg:
    Artists.append(i.text)
Artists
    
     #Upload_date
try:
    no_rtg_tg=driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[5]")
    Upload_date.append(no_rtg_tg.text)
except:
    Upload_date.append('-')


Upload_date=[]
for i in no_rtg_tg:
    Upload_date.append(i.text)
Upload_date    
    
    
    #views
try:
    price_tg=driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[4]")
    views.append(price_tg.text)
except:
    views.append('-')



views=[]
for i in price_tg:
    views.append(i.text)
views
               
        
    
print("Done Scraping!!!")
print(len(Ranks))
```
