```python
#Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).
#now we will send get request to the webpage server to get the source code of the page which is to be scraped by using the request library
page=requests.get("https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1") 
#Now we will parse this source code by using BeautifulSoup and the parser used here is html.parser as the source code is in HTML.
soup=BeautifulSoup(page.content)
#First, we will extract the html tag where we have the movies name.
name=soup.find_all("td",class_="titleColumn")
movies_name=[] #empty list
for i in name: 
    for j in i.find_all("a"):
        movies_name.append(j.text.replace("\n",""))
#movies_name
```

```python
page=requests.get("https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1") 
soup=BeautifulSoup(page.content)
name=soup.find_all("td",class_="titleColumn")
Year=[] #empty list
for i in name: 
    for j in i.find_all("span",class_="secondaryInfo"):
        Year.append(j.text.replace("\n",""))
#Year
```

```python
page=requests.get("https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1") 
soup=BeautifulSoup(page.content)
rate=soup.find_all("td",class_="ratingColumn imdbRating")
Rate=[] #empty list
for i in rate: 
    for j in i.find_all("strong"):
        Rate.append(j.text.replace("\n",""))
#Rate
```

```python
import pandas as pd
top_movies=pd.DataFrame({})
top_movies['Movies names']=movies_name[0:100]
top_movies['Year']=Year[0:100]
top_movies['Rate']=Rate[0:100]
top_movies
```

```python
#Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’
page=requests.get("https://bookpage.com/reviews") 
soup=BeautifulSoup(page.content)
name=soup.find_all("h4",class_="italic")
book_name=[] #empty list
for i in name: 
    for j in i.find_all("a"):
        book_name.append(j.text.replace("★",""))
```

```python
#books' author names
page=requests.get("https://bookpage.com/reviews") 
soup=BeautifulSoup(page.content)
name=soup.find_all("div",class_="flex-article-content")
author_name=[] #empty list
for i in name: 
    for j in i.find_all("p",class_="sans bold"):
        author_name.append(j.text.replace("\n",""))
     
```

```python
#books' genre
page=requests.get("https://bookpage.com/reviews") 
soup=BeautifulSoup(page.content)
name=soup.find_all("div",class_="flex-article-content")
genre=[] #empty list
for i in name: 
    for j in i.find_all("p", class_="genre-links hidden-phone"):
        genre.append(j.text.replace("\n",""))
```

```python
#books' review
page=requests.get("https://bookpage.com/reviews") 
soup=BeautifulSoup(page.content)
name=soup.find_all("div",class_="flex-article-content")
book_rev=[] #empty list
for i in name: 
    for j in i.find_all("p", class_="excerpt"):
        book_rev.append(j.text.replace("\n",""))
```

```python
import pandas as pd
Books=pd.DataFrame({})
Books['Book Name']=book_name[0:5]
Books['Author Name']=author_name[0:5]
Books['Genre']=genre[0:5]
Books['Book Review']=book_rev[0:5]
Books
```
